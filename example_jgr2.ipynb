```python
year=cpp.np.arange(1944,2050)
rage=cpp.np.arange(60,61)
cage=cpp.np.arange(65,66)
```

```python
cpp_out=pd.DataFrame(columns=['byear','retage','claimage','benefit','benefit_s1','benefit_s2','ampe', 'ampe_s1', 'ampe_s2'])
```

```python
count=0
for y in year:
    rules=cpp.rules(qpp=True)
    case = cpp.account(byear=y,rules=rules)
    for r in rage:
        for c in cage:
            yr18=np.max([case.gYear(18),rules.start])
            nbre_con =case.gYear(r)-yr18
            case.SetHistory_ratio(retage=r,a={"r":1.5,'n':nbre_con})
            case.SetHistory_fam(retage=r,age_birth=[30])
            case.RunCase(claimage=c)
            ncontrib=case.ncontrib
            ncontrib_s1=case.ncontrib_s1
            ncontrib_s2=case.ncontrib_s2
            ben=case.benefit
            ben_s1=case.benefit_s1
            ben_s2=case.benefit_s2
            ampe = case.ampe
            ampe_s1 = case.ampe_s1
            ampe_s2 = case.ampe_s2
            cpp_out.at[count,'byear']=y
            cpp_out.at[count,'retage']=r
            cpp_out.at[count,'claimage']=c
            cpp_out.at[count,'ncontrib']=ncontrib
            cpp_out.at[count,'ncontrib_s1']=ncontrib_s1
            cpp_out.at[count,'ncontrib_s2']=ncontrib_s2
            cpp_out.at[count,'benefit']=ben
            cpp_out.at[count,'benefit_s1']=ben_s1
            cpp_out.at[count,'benefit_s2']=ben_s2
            cpp_out.at[count,'ampe']=ampe
            cpp_out.at[count,'ampe_s1']=ampe_s1
            cpp_out.at[count,'ampe_s2']=ampe_s2
            count+=1
            case.ResetCase()    
```
