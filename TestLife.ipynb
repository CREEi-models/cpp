# Taux de mortalité par province (historique et futur)

On utilise les taux de mortalité historique provenant de la Base de donnée sur la longévité canadienne. On y colle les projections futurs de StatsCan

```python
from calculators import life
```

Le petit module (par default, Quebec, Hommes, scenario moyen pour long-terme)

```python
x = life.table()
```

Evolution des taux de mortalite par age

```python
%matplotlib inline
from matplotlib import pyplot as plt
ages = [30,40,50,60,70,80]
years = [i for i in range(1921,2063)]
plt.figure()
for age in ages:
    plt.plot(years,x.tab[str(age)],label=str(age))
plt.xlabel('year')
plt.ylabel('mortality rate')
plt.legend()
plt.show()
```

Taux de survie par cohorte d'un age a un autre:

```python
data = [x.sx(b,65,85) for b in range(1940,1980)]
byrs = [b for b in range(1940,1980)]
plt.figure()
plt.plot(byrs,data)
plt.xlabel('birth cohort')
plt.ylabel('probability survives to 85 given age 65')
plt.show()
```

Pour moi et David:

```python
ex =  [x.ex(b,40) for b in range(1940,1980)]
plt.figure()
plt.plot(byrs,ex)
plt.xlabel('birth year')
plt.ylabel('life expectancy at 40')
plt.show()
```

Plus sérieusement pour le RRQ:

```python
ex =  [x.ex(b,60) for b in range(1940,1980)]
plt.figure()
plt.plot(byrs,ex)
plt.xlabel('birth year')
plt.ylabel('life expectancy at 60')
plt.show()
```

Et finalement, le prix d'une rente a 60 ans (1$ de rente) avec différents taux de rendement

```python
ann1 =  [x.ann(b,60,0.03) for b in range(1940,1980)]
ann2 =  [x.ann(b,60,0.01) for b in range(1940,1980)]

plt.figure()
plt.plot(byrs,ann1,label='3 pct real return')
plt.plot(byrs,ann2,label='1 pct real return')
plt.xlabel('birth year')
plt.ylabel('price of annuity')
plt.legend()
plt.show()
```
