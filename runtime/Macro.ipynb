## Mortality Rates

```python
sc = pd.read_excel("params/quotients.xlsx",sheet_name='DTH_Tx_Scenario_M',skiprows=3)
sc.tail()
```

```python
qc = sc[sc.CGT==24]
qc = qc.drop(columns=[-1,'110+','CGT'])
names = list(qc.columns.values)
names[0] = 'year'
names[1] = 'gender'
qc.columns = names;
qc['year'], qc['temp'] = qc['year'].str.split('-', 1).str
qc = qc.drop(columns=['temp'])
qc.head()
```

We now can transform quotients into central death rates

```python
for a in range(0,110):
    qc[a] = 2.0*qc[a]/(2.0-qc[a])
```
